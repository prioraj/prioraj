```python
import matplotlib.pyplot as plt

# Plot a pixel (point)
def put_pixel(x, y, color="red"):
    plt.scatter(x, y, c=color, s=10)  # s = size of dot

# Plot 8-way symmetry
def draw_circle(xc, yc, x, y, color="red"):
    put_pixel(xc + x, yc + y, color)
    put_pixel(xc - x, yc + y, color)
    put_pixel(xc + x, yc - y, color)
    put_pixel(xc - x, yc - y, color)
    put_pixel(xc + y, yc + x, color)
    put_pixel(xc - y, yc + x, color)
    put_pixel(xc + y, yc - x, color)
    put_pixel(xc - y, yc - x, color)

# Bresenhamâ€™s Circle Algorithm (NO plt.show here!)
def circle_bres(xc, yc, r, color="red"):
    x, y = 0, r
    d = 3 - 2 * r

    while y >= x:
        draw_circle(xc, yc, x, y, color)

        if d <= 0:
            d += 4 * x + 6
        else:
            d += 4 * (x - y) + 10
            y -= 1
        x += 1
```

```python
plt.figure(figsize=(6,6))
for r in [5, 8, 12, 20]:
    circle_bres(50, 50, r, "blue")

plt.gca().set_aspect("equal")
plt.title("Task 1: Circles with different radii")
plt.show()
```

```python
plt.figure(figsize=(6,6))
xc, yc, r = 50, 50, 20
x, y, d = 0, r, 3 - 2 * r

while y >= x:
    for i in range(xc - x, xc + x + 1):
        put_pixel(i, yc + y, "cyan")
        put_pixel(i, yc - y, "cyan")
    for i in range(xc - y, xc + y + 1):
        put_pixel(i, yc + x, "cyan")
        put_pixel(i, yc - x, "cyan")

    if d <= 0:
        d += 4 * x + 6
    else:
        d += 4 * (x - y) + 10
        y -= 1
    x += 1

plt.gca().set_aspect("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,6))

r = 12   # radius
colors = ["blue", "black", "red", "yellow", "green"]

# Centers for intersecting Olympic rings
centers = [
    (0, 0),   # Blue
    (20, 0),   # Black (overlaps with Blue)
    (40, 0),  # Red (overlaps with Black)
    (10, -15),   # Yellow (overlaps with Blue & Black)
    (30, -15)   # Green (overlaps with Black & Red)
]

# Draw circles
for (xc, yc), c in zip(centers, colors):
    circle_bres(xc, yc, r, c)

plt.gca().set_aspect("equal")
plt.title("Task 7: Olympic Rings (Intersecting)")
plt.show()
```
