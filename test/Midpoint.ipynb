```python
import matplotlib.pyplot as plt

def midpoint_line_with_reference(x1, y1, x2, y2, color='blue'):
    x_points = []
    y_points = []

    dx = x2 - x1
    dy = y2 - y1

    step_x = 1 if dx >= 0 else -1
    step_y = 1 if dy >= 0 else -1
    dx = abs(dx)
    dy = abs(dy)

    x, y = x1, y1

    if dx >= dy:  # slope <= 1
        p = 2 * dy - dx
        while x != x2:
            x_points.append(x)
            y_points.append(y)
            x += step_x
            if p < 0:
                p += 2 * dy
            else:
                y += step_y
                p += 2 * (dy - dx)
    else:  # slope > 1
        p = 2 * dx - dy
        while y != y2:
            x_points.append(x)
            y_points.append(y)
            y += step_y
            if p < 0:
                p += 2 * dx
            else:
                x += step_x
                p += 2 * (dx - dy)

    x_points.append(x2)
    y_points.append(y2)

    plt.plot(x_points, y_points, marker='o', color=color, label='Midpoint Algorithm Points')
    plt.plot([x1, x2], [y1, y2], color='red', linestyle='--', label='Ideal Straight Line')

    plt.title(f"Midpoint Line ({x1},{y1}) to ({x2},{y2})")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.legend()
    plt.show()

# Get user inputs
x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

midpoint_line_with_reference(x1, y1, x2, y2, color='purple')
```
